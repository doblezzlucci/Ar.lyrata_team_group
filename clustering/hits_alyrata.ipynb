```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hight=pd.read_csv("blast_strict.tsv", sep=r"\s+", header=None)
medium=pd.read_csv("blast_medium.tsv", sep=r"\s+", header=None)
low=pd.read_csv("blast_lenient.tsv", sep=r"\s+", header=None)
```

```python
hight.head()
```

```python
hight.columns = medium.columns = low.columns = [
    "query_id", "subject_id", "pident", "length",
    "mismatch", "gapopen", "qstart", "qend",
    "sstart", "send", "evalue", "bitscore",
    "qlen", "slen"
]
```

```python
# Histogram of bitscore
plt.figure(figsize=(8,5))
plt.hist(hight["bitscore"], bins=40, color="steelblue", edgecolor="black", alpha=0.7)
plt.xlabel("Bitscore")
plt.ylabel("Number of hits")
plt.title("Distribution of BLAST bitscores of stringent hits")
plt.show()
```

```python
# Histogram of bitscore
plt.figure(figsize=(8,5))
plt.hist(medium["bitscore"], bins=40, color="steelblue", edgecolor="black", alpha=0.7)
plt.xlabel("Bitscore")
plt.ylabel("Number of hits")
plt.title("Distribution of BLAST bitscores of middle hits")
plt.show()
```

```python
# Histogram of bitscore
plt.figure(figsize=(8,5))
plt.hist(low["bitscore"], bins=40, color="steelblue", edgecolor="black", alpha=0.7)
plt.xlabel("Bitscore")
plt.ylabel("Number of hits")
plt.title("Distribution of BLAST bitscores of lenient hits")
plt.show()
```
